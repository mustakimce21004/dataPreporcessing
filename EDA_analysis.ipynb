```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dataset.csv")
```

```python

display(data.head())
```

```python
print("\n--- Dataset Information ---")
print(data.info())

print("\n--- Summary Statistics ---")
print(data.describe(include='all'))

print("\n--- First Few Rows ---")
print(data.head())
```

```python
# Convert 'Age' and 'Salary' to numeric, coercing errors
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')


# Distribution of numerical columns
plt.figure(figsize=(10,5))
sns.histplot(data['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(data['Salary'], kde=True)
plt.title("Salary Distribution")
plt.show()

# Boxplots to detect outliers
plt.figure(figsize=(6,4))
sns.boxplot(x='Age', data=data)
plt.title("Boxplot of Age")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='Salary', data=data)
plt.title("Boxplot of Salary")
plt.show()

# Pairplot (if numeric columns exist)
sns.pairplot(data, hue="purchased")
plt.suptitle("Pairplot of Dataset", y=1.02)
plt.show()

# Heatmap for correlation (only numeric data)
plt.figure(figsize=(6,4))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
# Step 5: Interpretation (to be done manually)
print("\n--- Observations ---")
print(" Age and Salary show variation across countries.")
print(" Missing values detected in 'Age' and 'Salary'.")
print(" Correlation heatmap shows how numeric features relate.")

# Step 6: Data Preprocessing (Handle missing values, clean text)
# Fill missing Age with mean, Salary with median
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Salary'].fillna(data['Salary'].median(), inplace=True)

# Standardize text cases in 'country' and 'purchased'
data['country'] = data['country'].str.title()
data['purchased'] = data['purchased'].str.capitalize()

print("\n--- Preprocessed Data ---")
print(data)
```
